```python
def gauss (x, sig, mu):
    gaussian = (2*np.pi)**(-7.0/2)*np.linalg.det(sig)**(-1.0/2)*np.exp(-0.5*np.matrix(x-mu) @ np.linalg.inv(sig) @ np.transpose(np.matrix(x-mu)))
    return float(gaussian)
```

```python
def kmeansLikelihood (test, train, means, cluster0, cluster1, cluster2, cluster3, cluster4, cluster5, cluster6,  clusterNum):
    #find pi values
    pi0 = cluster0.shape[0]/train.shape[0]
    pi1 = cluster1.shape[0]/train.shape[0]
    pi2 = cluster2.shape[0]/train.shape[0]
    pi3 = cluster3.shape[0]/train.shape[0]
    pi4 = cluster4.shape[0]/train.shape[0]
    pi5 = cluster5.shape[0]/train.shape[0]
    pi6 = cluster6.shape[0]/train.shape[0]
    
    means0 = np.zeros(6)
    means1 = np.zeros(6)
    means2 = np.zeros(6)
    means3 = np.zeros(6)
    means4 = np.zeros(6)
    means5 = np.zeros(6)
    means6 = np.zeros(6)
    
    x = np.ones(6)
    covariance0 = np.diagflat(x)
    covariance1 = np.diagflat(x)
    covariance2 = np.diagflat(x)
    covariance3 = np.diagflat(x)
    covariance4 = np.diagflat(x)
    covariance5 = np.diagflat(x)
    covariance6 = np.diagflat(x)
    covariance7 = np.diagflat(x)
    covariance8 = np.diagflat(x)
    covariance9 = np.diagflat(x)
    covariance10 = np.diagflat(x)
    covariance11 = np.diagflat(x)
    covariance12 = np.diagflat(x)
    
    if clusterNum>0:
        covariance0 = cluster0.cov()
        means0 = means[0]
    if clusterNum>1:
        covariance1 = cluster1.cov()
        means1 = means[1]
    if clusterNum>2:
        covariance2 = cluster2.cov()
        means2 = means[2]
    if clusterNum>3:
        covariance3 = cluster3.cov()
        means3 = means[3]
    if clusterNum>4:
        covariance4 = cluster4.cov()
        means4 = means[4]
    if clusterNum>5:
        covariance5 = cluster5.cov()
        means5 = means[5]
    if clusterNum>6:
        covariance6 = cluster6.cov()
        means6 = means[6]
    

    clustSum = np.zeros(test.shape[0])
    for i in range(test.shape[0]):
        x = test[i:i+1]
        clustSum[i] = np.log((pi0*gauss(x, covariance0, means0 ) + pi1*gauss(x, covariance1, means1 ) + pi2*gauss(x, covariance2, means2 ) + pi3*gauss(x, covariance3, means3 ) + pi4*gauss(x, covariance4, means4 ) + pi5*gauss(x, covariance5, means5 ) + pi6*gauss(x, covariance6, means6 )))
    mle = np.sum(clustSum)
    return mle
 
```

```python
mles = np.zeros(10)
predicted = np.zeros((2200))
actual = np.zeros((2200))
start = 0
counter = 0
digitMle = np.shape((10))
for i in (range(10)):
    for j in tqdm(range(220)):
        while not (np.isnan(test_df.iloc[counter][0])):
            if i == 9 and j == 219:
                counter = test_df.shape[0]
                break;
            counter=counter+1
        testing = test_df.iloc[start:counter].values
        mle0 = kmeansLikelihood (testing, zero, means0, cluster00, cluster01, cluster02, cluster03, cluster04, cluster05, cluster06, zeroClusterNum)
        mle1 = kmeansLikelihood (testing, one, means1, cluster10, cluster11, cluster12, cluster13, cluster14, cluster15, cluster16, oneClusterNum)
        mle2 = kmeansLikelihood (testing, two, means2, cluster20, cluster21, cluster22, cluster23, cluster24, cluster25, cluster26, twoClusterNum)
        mle3 = kmeansLikelihood (testing, three, means3, cluster30, cluster31, cluster32, cluster33, cluster34, cluster35, cluster36, threeClusterNum)
        mle4 = kmeansLikelihood (testing, four, means4, cluster40, cluster41, cluster42, cluster43, cluster44, cluster45, cluster46, fourClusterNum)
        mle5 = kmeansLikelihood (testing, five, means5, cluster50, cluster51, cluster52, cluster53, cluster54, cluster55, cluster56, fiveClusterNum)
        mle6 = kmeansLikelihood (testing, six, means6, cluster60, cluster61, cluster62, cluster63, cluster64, cluster65, cluster66, sixClusterNum)
        mle7 = kmeansLikelihood (testing, seven, means7, cluster70, cluster71, cluster72, cluster73, cluster74, cluster75, cluster76, sevenClusterNum)
        mle8 = kmeansLikelihood (testing, eight, means8, cluster80, cluster81, cluster82, cluster83, cluster84, cluster85, cluster86, eightClusterNum)
        mle9 = kmeansLikelihood (testing, nine, means9, cluster90, cluster91, cluster92, cluster93, cluster94, cluster95, cluster96, nineClusterNum)
        likelihoods = [mle0, mle1, mle2, mle3, mle4, mle5, mle6, mle7, mle8, mle9]
        digit = np.argmax(likelihoods)
        actual[i*220+j] = i
        predicted[i*220+j] = digit
        counter = counter + 1
        start = counter

        
```
